```python
import pandas as pd

df = pd.read_csv('station_weather_data.csv')
df = df.drop(['PRCP_ATTRIBUTES','TAVG_ATTRIBUTES','TMAX_ATTRIBUTES','TMIN_ATTRIBUTES'],axis=1)
df = df.fillna(0)
df.groupby('NAME').mean()
```

```python
import numpy as np
def get_energy_solution(row):
    if row['TMAX'] > 40 and row['TAVG'] > 25 and row['TMIN'] > 20 and row['PRCP'] < 10:
        return 'solar'
    elif row['PRCP'] > 50 and row['TMAX'] < 25 and row['TAVG'] < 20 and row['TMIN'] < 15:
        return 'hydro'
    else:
        return 'wind'
df['energy_solution'] = df.apply(get_energy_solution, axis=1)

# print the sorted dataframe
print(df.sort_values(by=['energy_solution', 'TAVG'], ascending=[True, False]))
```

```python
df.groupby('energy_solution').mean()
```
